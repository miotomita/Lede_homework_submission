# Apple cultivars

* **Data:** `apples.csv`
* **Description:** Types of apples
* **Source:** https://en.wikipedia.org/wiki/List_of_apple_cultivars
* **Columns of interest:**
    * `Common name` - the non-scientific name of the apple cultivar
    * `Origin` - where it came from
    * `First developed` - year it was developed
    * `Comment` - Description
    * `Use` - whether best for eating, baking, etc

Note that you'll need to brush up on `.str.contains` for this notebook.

## Read in your data

```python
url = "https://en.wikipedia.org/wiki/List_of_apple_cultivars"
```

```python
#Table of apples
table = pd.read_html(url)[1]
```

```python
table = table.replace("\[.*\]","",regex=True)
```

```python
table = table.drop('Image',axis=1)
```

```python
table.columns = table.columns.str.lower().str.replace(" ","_")
```

```python
table.sample(10)
```

## Clean up the "First developed" column

Sometimes it's a year and sometimes it is an "introduced" year, like `Introduced 1826`. We just want all of them to be years, so feel free to remove the "Introduced" part.

```python
#this code returns rows with multiple years like, "19812009"
#table.first_developed = table.first_developed.str.replace("\D","", regex=True)
```

```python
#rows with multiple years
table[table.first_developed.str.contains('\d\d\d.*\d\d\d\d$',na=False)]
```

```python
#rows with na values
table[table.first_developed.isna()]
```

```python
table.first_developed = table.first_developed.str.extract('.*(\d\d\d\d)').dropna().astype(int)
```

```python
table.head()
```

## What percentage of apples are just for eating?

```python
table.use.unique()
```

```python
table.use = table.use.str.lower().str.strip()
```

```python
#52% of all, including na values, are just for eating  
table.use.value_counts(normalize=True,dropna=False).eating.round(2)
```

## Apples are from Kazakhstan, but they've spread since ancient times: calculate the top seven countries for apple cultivars in this dataset

```python
table.origin = table.origin.str.extract('[\.\,\?]*(\w* *\w+ *\w+$)')
```

```python
table.origin = table.origin.str.strip().str.lower()
```

```python
#cleaning up
table.origin = table.origin.replace({
    "united states":"us",
    "usa":"us",
    "louisiana":"us",
    "new york":"us",
    "idaho":"us",
    "connecticut":"us",
    "washington":"us",
    "florida":"us",
    "netherland":"netherlands",
    "virginia":"us",
    "massachusetts":"us",
    "oregon":"us",
    "arkansas":"us",
    "california":"us",
    "new jersey":"us",
    "michigan":"us",
    "new hampshire":"us",
    "geneva new york":"us",
    "holland":"netherlands",
    "united kingdom":"uk",
    "england":"uk",
    "wales":"uk",
    "or lincolnshire england":"uk",
    "berkshire england":"uk",
    "wocestershire england":"uk",
    "nova scotia":"canada",
    "ontario":"canada",
    "manitoba canada":"canada",
    "tyrol":"australia",
    "new zeeland":"new zealand",
    "east germany":"germany",
    "saxonia germany":"germany"
    
})
```

```python
table.loc[table.origin.str.contains(" us$",na=False), "origin"] = "us"
```

```python
table.loc[table.origin=='1908',["origin","first_developed"]] = [np.nan, 1908]
```

## How old are these apple cultivars?

You could use `.describe()`, but I want a graph. **What kind of chart do you use?** I want to see a graph that shows me a summary when these cultivars were discovered/introduced, not a graph of each individual one's year of introduction.

```python
table['age'] = datetime.now().year - table.first_developed
```

```python
table.age.describe()
```

```python
table.age.hist(bins=20, ec="white", range=(0, 500))
plt.xlabel("age")
plt.ylabel("number of apple cultivars")
plt.title("The age of apple cultivars")
plt.show()
```

## How old are most apple cultivars that are used for creating cider?

Cider used to be a big deal, so I would assume they're older. Include **anything that includes "Cider" anywhere in the `Use` description**. Make a graph, same as the last question.

```python
ciders = table[table.use.str.contains('cider', na=False)]
```

```python
ciders.age.describe()
```

```python
ciders.age.hist(bins=20, ec="white", range=(0, 500))
plt.xlabel("age")
plt.ylabel("number of apple cultivars")
plt.title("The age of apple cultivars used for creating cider")
plt.show()
```
